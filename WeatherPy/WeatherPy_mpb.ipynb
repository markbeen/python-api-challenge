```python
## I downloaded the json file from openweather, and then created a dataframe from it:
cityList = pd.read_json("current.city.list.json")
len(cityList)
```

```python
## Randomly sample 500 cities from the json file:
sampleCity = cityList.sample(n=500)
len(sampleCity)
```

```python
names = list(sampleCity.columns) 
countryNames = sampleCity['country']
cityNames = sampleCity['name']
```

```python
cityData = {'City': cityNames, 'Country': countryNames}
cityWeather = pd.DataFrame(data=cityData)
cityWeather.reset_index(drop=True, inplace=True)
cityWeather.head()
```

```python
## append desired weather parameters as new series to city dataframe
cityWeather['Temperature'] = ""
cityWeather['Humidity'] = ""
cityWeather['Cloudiness'] = ""
cityWeather['Wind Speed'] = ""
cityWeather['Latitude'] = ""
cityWeather['Longitude'] = ""
cityWeather.head()
```

```python
## define the location of the weather api (remember: the ?q means that you can insert a variable)
weather_url = "http://api.openweathermap.org/data/2.5/weather?"
#weather_url = "http://api.openweathermap.org/data/2.5/weather?"           

# define desired weather units
units = 'imperial'
```

## make for loop to extract weather data for each city
record = 0
for index, row in cityWeather.iterrows():
    theCity = row['City']
    theCountry = row['Country']
    query_url = weather_url + "appid=" + api_key + '&units=' + units + "&q=" + theCity + ',' + theCountry
    try:
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
        latitude = weather_json["coord"]["lat"]
        longitude = weather_json["coord"]["lon"]
        temp = weather_json["main"]["temp"]
        humidity = weather_json["main"]["humidity"]
        cloud = weather_json["clouds"]["all"]
        wind = weather_json["wind"]["speed"]
        cityWeather.loc[index,"Temperature"] = temp
        cityWeather.loc[index, "Humidity"]= humidity
        cityWeather.loc[index, "Wind Speed"] = wind
        cityWeather.loc[index, "Cloudiness"] = cloud
        cityWeather.loc[index,"Latitude"] = latitude
        cityWeather.loc[index, "Longitude"] =longitude
        print(query_url)
    except:
        print('error')
    record += 1
    
cityWeather.to_csv('cityWeatherData.csv')

```python
## make two dataframes, 1 for northern lats and 1 for southern lats
northern = cityWeather.loc[cityWeather['Latitude'] > 0]
southern = cityWeather.loc[cityWeather['Latitude'] < 0]
```

```python
## get data out of northern dataframe
northernTemps = northern.Temperature
northernLats = northern.Latitude
northernLongs = northern.Longitude
northernHums = northern.Humidity
northernWinds = northern['Wind Speed']
northernClouds = northern.Cloudiness
```

```python
## get data out of southern dataframe
southernTemps = southern.Temperature
southernLats = southern.Latitude
southernLongs = southern.Longitude
southernHums = southern.Humidity
southernWinds = southern['Wind Speed']
southernClouds = southern.Cloudiness
```

```python
## plot nothern latitudes versus northern temperatures
plt.scatter(northernLats, northernTemps) ;
plotTitle = "Northern Latitudes VS Temperature"
plt.title(plotTitle);
plt.xlabel("Latitude");
plt.ylabel("Temperature (F)");

# add regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(northernLats, northernTemps)
regress_values = northernLats * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(northernLats,regress_values,"r-")
rValue, pValue = pearsonr(northernLats, northernTemps)
plt.text(0, 35,f'The Pearons R Value is {round(rValue, 3)}', color='red') ;
plt.savefig(plotTitle)
```
